```python
import sqlite3
import pandas as pd

csv_filename = '/content/lung_cancer_Dataset (1).csv'


df = pd.read_csv(csv_filename)
conn = sqlite3.connect(':memory:')
df.to_sql('lung_cancer', conn, if_exists='replace', index=False)

df.head()
```

**Checking for NULL values**

```python
query = """
SELECT *
FROM lung_cancer
WHERE Smoker IS NULL OR
      Years_of_Smoking IS NULL OR
      Cigarettes_per_Day IS NULL OR
      Passive_Smoker IS NULL OR
      Family_History IS NULL OR
      Lung_Cancer_Diagnosis IS NULL OR
      Cancer_Stage IS NULL OR
      Survival_Years IS NULL OR
      Adenocarcinoma_Type IS NULL OR
      Air_Pollution_Exposure IS NULL OR
      Occupational_Exposure IS NULL OR
      Indoor_Pollution IS NULL OR
      Healthcare_Access IS NULL OR
      Early_Detection IS NULL OR
      Treatment_Type IS NULL OR
      Developed_or_Developing IS NULL OR
      Annual_Lung_Cancer_Deaths IS NULL OR
      Lung_Cancer_Prevalence_Rate IS NULL OR
      Mortality_Rate IS NULL
"""
null_values_df = pd.read_sql_query(query, conn)
null_values_df
```

**Delete Null Rows**

```python
delete_query = """
DELETE FROM lung_cancer
WHERE Smoker IS NULL OR
      Years_of_Smoking IS NULL OR
      Cigarettes_per_Day IS NULL OR
      Passive_Smoker IS NULL OR
      Family_History IS NULL OR
      Lung_Cancer_Diagnosis IS NULL OR
      Cancer_Stage IS NULL OR
      Survival_Years IS NULL OR
      Adenocarcinoma_Type IS NULL OR
      Air_Pollution_Exposure IS NULL OR
      Occupational_Exposure IS NULL OR
      Indoor_Pollution IS NULL OR
      Healthcare_Access IS NULL OR
      Early_Detection IS NULL OR
      Treatment_Type IS NULL OR
      Developed_or_Developing IS NULL OR
      Annual_Lung_Cancer_Deaths IS NULL OR
      Lung_Cancer_Prevalence_Rate IS NULL OR
      Mortality_Rate IS NULL
"""
# Execute the deletion query
conn.execute(delete_query)
conn.commit()
```

# ▶**Basic Level Queries**

---



**1. Retrieve all records for individuals diagnosed with lung cancer.**

```python
query = "SELECT * FROM lung_cancer WHERE Lung_Cancer_Diagnosis = 'Yes' "
pd.read_sql_query(query, conn)
```

**2. Count the number of smokers and non-smokers.**

```python
query = "SELECT smoker, COUNT(*) AS count FROM lung_cancer GROUP BY smoker"
pd.read_sql_query(query, conn)
```

**3. List all unique cancer stages present in the dataset.**

```python
query="select distinct cancer_stage from lung_cancer"
pd.read_sql_query(query,conn)
```

**4. Retrieve the average number of cigarettes smoked per day by smokers.**

```python
query = "SELECT AVG(cigarettes_per_day) AS avg_cigarettes FROM lung_cancer WHERE smoker = 'Yes'"
pd.read_sql_query(query, conn)
```

**5. Count the number of people exposed to high air pollution.**

```python
query = "SELECT COUNT(*) AS high_pollution_count FROM lung_cancer WHERE Air_Pollution_Exposure = 'High'"
pd.read_sql_query(query, conn)
```

**6. Find the top 5 countries with the highest lung cancer deaths.**

```python
query = """
SELECT Country, SUM(Annual_Lung_Cancer_Deaths) AS total_deaths
FROM lung_cancer
GROUP BY Country
ORDER BY total_deaths DESC
LIMIT 5
"""
pd.read_sql_query(query, conn)
```

**7. Count the number of people diagnosed with lung cancer by gender**

```python
query = "SELECT Gender, COUNT(*) AS count FROM lung_cancer WHERE Lung_Cancer_Diagnosis = 'Yes' GROUP BY Gender"
pd.read_sql_query(query, conn)
```

**8. Retrieve records of individuals older than 60 who are diagnosed with lung cancer**

```python
query = "SELECT * FROM lung_cancer WHERE Age > 60 AND Lung_Cancer_Diagnosis = 'Yes'"
pd.read_sql_query(query, conn)
```

# ▶**`Intermediate Level Queries`**

**1. Find the percentage of smokers who developed lung cancer**


```python
query = """
SELECT
    (COUNT(CASE WHEN Smoker = 'Yes' AND Lung_Cancer_Diagnosis = 'Yes' THEN 1 END) * 100.0) /
    COUNT(CASE WHEN Smoker = 'Yes' THEN 1 END) AS smoker_lung_cancer_percentage
FROM lung_cancer
"""
pd.read_sql_query(query, conn)
```

**2. Calculate the average survival years based on cancer stages**

```python
query = "SELECT Cancer_Stage, AVG(Survival_Years) AS avg_survival FROM lung_cancer GROUP BY Cancer_Stage"
pd.read_sql_query(query, conn)
```

**3. Count the number of lung cancer patients based on passive *smoking***

```python
query = "SELECT Passive_Smoker, COUNT(*) AS count FROM lung_cancer WHERE Lung_Cancer_Diagnosis = 'Yes' GROUP BY Passive_Smoker"
pd.read_sql_query(query, conn)
```

**4. Find the country with the highest lung cancer prevalence rate**

```python
query = "SELECT Country, MAX(Lung_Cancer_Prevalence_Rate) AS max_prevalence FROM lung_cancer"
pd.read_sql_query(query, conn)
```

**5. Identify the smoking years' impact on lung cancer**

```python
query = """
SELECT Years_of_Smoking, COUNT(*) AS lung_cancer_cases
FROM lung_cancer
WHERE Lung_Cancer_Diagnosis = 'Yes'
GROUP BY Years_of_Smoking
ORDER BY Years_of_Smoking DESC
"""
pd.read_sql_query(query, conn)
```

**6. Determine the mortality rate for patients with and without early detection**

```python
query = """
SELECT Early_Detection,
    (COUNT(CASE WHEN Mortality_Rate > 0 THEN 1 END) * 100.0) / COUNT(*) AS mortality_rate
FROM lung_cancer
GROUP BY Early_Detection
"""
pd.read_sql_query(query, conn)
```

**7. Group lung cancer prevalence rate by developed vs. developing countries**

```python
query = "SELECT Developed_or_Developing, AVG(Lung_Cancer_Prevalence_Rate) AS avg_prevalence FROM lung_cancer GROUP BY Developed_or_Developing"
pd.read_sql_query(query, conn)
```

# ▶**Advanced Level Queries**


**1. Correlation between lung cancer prevalence and air pollution levels.**

```python
query = "SELECT Air_Pollution_Exposure, AVG(Lung_Cancer_Prevalence_Rate) AS avg_prevalence FROM lung_cancer GROUP BY Air_Pollution_Exposure"
pd.read_sql_query(query, conn)
```

**2. Average age of lung cancer patients for each country.**

```python
query = "SELECT Country, AVG(Age) AS avg_age FROM lung_cancer WHERE Lung_Cancer_Diagnosis = 'Yes' GROUP BY Country"
pd.read_sql_query(query, conn)
```

**3. Risk factor of lung cancer by smoker status, passive smoking, and family history**

```python
query = """
SELECT Smoker, Passive_Smoker, Family_History, COUNT(*) AS lung_cancer_cases
FROM lung_cancer
WHERE Lung_Cancer_Diagnosis = 'Yes'
GROUP BY Smoker, Passive_Smoker, Family_History
ORDER BY lung_cancer_cases DESC
"""
pd.read_sql_query(query, conn)
```

**4. Rank countries based on mortality rate**


```python
query = """
SELECT Country, AVG(Mortality_Rate) AS avg_mortality_rate
FROM lung_cancer
GROUP BY Country
ORDER BY avg_mortality_rate DESC
"""
pd.read_sql_query(query, conn)
```

**5. Determine if treatment type has a significant impact on survival years**

```python
query = "SELECT Treatment_Type, AVG(Survival_Years) AS avg_survival FROM lung_cancer GROUP BY Treatment_Type"
pd.read_sql_query(query, conn)
```

**6. Compare lung cancer prevalence in men vs. women across countries.**

```python
query = "SELECT Country, Gender, COUNT(*) AS lung_cancer_cases FROM lung_cancer WHERE Lung_Cancer_Diagnosis = 'Yes' GROUP BY Country, Gender"
pd.read_sql_query(query, conn)
```

**7. How occupational exposure, smoking, and air pollution collectively impact lung cancer rates**

```python
query = """
SELECT Occupational_Exposure, Smoker, Air_Pollution_Exposure, COUNT(*) AS lung_cancer_cases
FROM lung_cancer
WHERE Lung_Cancer_Diagnosis = 'Yes'
GROUP BY Occupational_Exposure, Smoker, Air_Pollution_Exposure
ORDER BY lung_cancer_cases DESC
"""
pd.read_sql_query(query, conn)
```

**8. Impact of early detection on survival years**


```python
query = "SELECT Early_Detection, AVG(Survival_Years) AS avg_survival FROM lung_cancer GROUP BY Early_Detection"
pd.read_sql_query(query, conn)
```
